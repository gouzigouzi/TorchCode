# 🟡 Solution: Implement BatchNorm

Reference solution for Batch Normalization (training mode).

```python
import torch
```

```python
# ✅ SOLUTION

def my_batch_norm(x, gamma, beta, eps=1e-5):
    mean = x.mean(dim=0)
    var = x.var(dim=0, unbiased=False)
    x_norm = (x - mean) / torch.sqrt(var + eps)
    return gamma * x_norm + beta
```

```python
# Verify
x = torch.randn(8, 4)
out = my_batch_norm(x, torch.ones(4), torch.zeros(4))
print("Column means:", out.mean(dim=0))
print("Column stds: ", out.std(dim=0))
```
