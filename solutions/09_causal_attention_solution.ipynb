# 🔴 Solution: Causal Self-Attention

Reference solution — softmax attention with an upper-triangular mask.

```python
import torch
import math
```

```python
# ✅ SOLUTION

def causal_attention(Q, K, V):
    d_k = K.size(-1)
    scores = torch.bmm(Q, K.transpose(1, 2)) / math.sqrt(d_k)  # [B, sq, sk]
    S = scores.size(-1)  # Sequence length (sk)
    mask = torch.triu(torch.ones(S, S, device=scores.device, dtype=torch.bool), diagonal=1)  # Upper triangular mask
    scores = scores.masked_fill(mask.unsqueeze(0), float('-inf'))
    weights = torch.softmax(scores, dim=-1)  # [B, sq, sk]
    return torch.bmm(weights, V)  # [B, sq, d_v]
```

```python
# Verify
torch.manual_seed(0)
Q = torch.randn(1, 4, 8)
K = torch.randn(1, 4, 8)
V = torch.randn(1, 4, 8)
out = causal_attention(Q, K, V)
print("Pos 0 == V[0]?", torch.allclose(out[:, 0], V[:, 0], atol=1e-5))
```
